```python
y = melbourne_data['SalePrice']
y = melbourne_data.SalePrice
```

```python
features = ['LotArea', 'Neighborhood', 'HouseStyle', 'YearBuilt', 'RoofStyle', 'BedroomAbvGr', 'GarageArea','Fence' ]
X = melbourne_data[features]
X.describe()
```

```python
X.head()
```

Defining the model whether it will be a random decision tree or otherwise and passing the parameters id available


```python
from sklearn.tree import DecisionTreeRegressor

melbourne_model = DecisionTreeRegressor(random_state=1)
```

We now fit the model to capture the details or features earlier provided

```python
from sklearn.preprocessing import LabelEncoder

#identify non-numeric
non_numeric_columns = [col for col in X.columns if X[col].dtype == "object"]

#encode non_numeric columns
encoder = LabelEncoder()
for col in non_numeric_columns:
  X.loc[:, col] = encoder.fit_transform(X[col])
#fit the model
melbourne_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses")
print(X.head())
print("The predictions are: ")
print(melbourne_model.predict(X.head()))
```
